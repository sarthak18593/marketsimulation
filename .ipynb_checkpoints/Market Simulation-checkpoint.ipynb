# Market Simulation 

## Legend :


* Q = QUOTE: Message that contains information on the event that occurred in the book
* T = TRADE: Message that contains information about a trade that occurred
* M = TRADE MATCH ELEMENT: Message that contains information on a match of a trade event
* V = VOLUME: Message that updates volume data after a trade event
* ST = DAILY STATISTIC: Message that provides information about a complete session
     * Cleared Volume
     * Fixing Price
     * Open Interest
     * Settlement Price
* LM = LIMITS DAILY: Message that provides the daily limits for the current session
* DS = SESSION STATISTICS: Message that provides information about the session during the session
     * High bid and Low ask
     * Indicative opening
     * Opening price
     * Session High and Low
* SS = SECURITY STATUS MESSAGE: Message that provides the security group market state change
* SC = SUMMARY COUNTS MESSAGE: Message that provides the number of matches from a trade event
* DH = DEPTH HEADER: Message that provides what changed in this depth update
